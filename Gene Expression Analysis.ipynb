#### PCA and T-SNE to differentiate control samples from test samples

```python
from sklearn.decomposition import PCA
```

##### In gene expression data:
- Rows often represent genes.
- Columns often represent samples.
    
##### Requirement for PCA
- PCA expects the input data matrix to have:
- Rows as samples.
- Columns as features.

That is why I transpose. The same convention applies to T-SNE

```python
pca = PCA(n_components=2)
dm_pca = pca.fit_transform(dm_data.T)
dm_pca
```

#### Differential Expression Analysis

```python
from scipy.stats import ttest_ind
```

```python
dm_data_T = dm_data.T
rn_data_T = rn_data.T
```

```python
dm_data_T['condition'] = dm_labels
dm_data_T['condition'] = dm_data_T['condition'].astype('category').cat.codes
```

```python
rn_data_T['condition'] = rn_labels
rn_data_T['condition'] = rn_data_T['condition'].astype('category').cat.codes
```

```python
dm_control = dm_data_T[dm_data_T['condition'] == 0]
dm_test = dm_data_T[dm_data_T['condition'] == 1]
```

```python
rn_control = rn_data_T[rn_data_T['condition'] == 0]
rn_test = rn_data_T[rn_data_T['condition'] == 1]
```

##### For Drosophila

```python
dm_p_values = []
for gene in dm_data_T.columns[:-1]:
    t_stat, p_val = ttest_ind(dm_control[gene], dm_test[gene])
    dm_p_values.append(p_val)

# Adjusting p-values for multiple testing (e.g., using Benjamini-Hochberg)
from statsmodels.stats.multitest import multipletests
adj_dm_p_values = multipletests(dm_p_values, method='fdr_bh')[1]

# Identifying differentially expressed genes
dm_diff_expressed_genes = dm_data_T.columns[:-1][adj_dm_p_values < 0.05]
print(dm_diff_expressed_genes)
```

```python
'CG6295' in dm_diff_expressed_genes
```

###### The CG6295 gene is supposedly not differentially expressed. It is the Drosophila ortholog of the mammalian Liph
